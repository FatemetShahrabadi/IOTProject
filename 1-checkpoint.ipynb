```python
import numpy as np
import matplotlib.pyplot as plt

######################
# Create perceptron
#####################
class NeuralNetwork():
    
    def __init__(self):
        # seeding for random number generation
        np.random.seed(1)
        self.weights = 2 * np.random.random((3, 1)) - 1

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def train(self, training_inputs, training_outputs, iterations):
        for iteration in range(iterations):
            output = self.calculate(training_inputs)
            #computing error rate
            error = training_outputs - output
            #performing weight adjustments
            adjustments = np.dot(training_inputs.T, error * self.sigmoid_derivative(output))
            self.weights += adjustments
            # classification in final step
            if(iteration == iterations-1):
                for i in range(len(output)):
                    if (output[i]<0.5):
                        # print(training_inputs[i])
                        predictN.append(training_inputs[i])
                    if (output[i]>0.5):
                        # print(training_inputs[i])
                        predictP.append(training_inputs[i])
                
              
    def calculate(self, inputs):
        inputs = inputs.astype(float)
        output = self.sigmoid(np.dot(inputs, self.weights))
        return output
    
# function for drawing class seprator
def abline(slope, intercept):
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
    

######################
# Create random data
#####################
my_inputs = []
my_outputs = []
predictN = []
predictP = []
random_positive = np.random.normal(100,45,100)
random_negative = np.random.normal(-100,45,100)
N=[[x,y,1] for x,y in zip(random_positive,random_negative)]
P=[[x,y,1] for x,y in zip(random_negative,random_positive)]
N1=[[x,y] for x,y in zip(random_positive,random_negative)]
P1=[[x,y] for x,y in zip(random_negative,random_positive)]
my_inputs.extend(N)
my_inputs.extend(P)
for x in range(0,100):
  my_outputs.append(0)
for x in range(100,200):
 my_outputs.append(1)

########################
# Data Visualization
########################
print('before classification')
plt.plot([x for x,y in N1], [y for x,y in N1], 'ro', label='N')
plt.plot([x for x,y in P1], [y for x,y in P1], 'go', label='P')

plt.title('synthetic linearly separable data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right')
plt.show()

######################
# Start classification with perceptron
#####################
if __name__ == "__main__":

    #initializing the neuron class
    neural_network = NeuralNetwork()
    training_inputs = np.array(my_inputs)
    training_outputs = np.array([my_outputs]).T
    neural_network.train(training_inputs, training_outputs, 15000)

    print("Ending Weights After Training: ")
    print(neural_network.weights)

   

########################
# Data Visualization
########################
plt.plot([x for x,y,z in predictN], [y for x,y,z in predictN], 'yo', label='N')
plt.plot([x for x,y,z in predictP], [y for x,y,z in predictP], 'bo', label='P')
print('after classification')
plt.title('synthetic linearly separable data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right')
abline(-neural_network.weights[0]/neural_network.weights[1],neural_network.weights[2])
plt.show()


```
